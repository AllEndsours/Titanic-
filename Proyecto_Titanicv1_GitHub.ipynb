4- Limpieza de datos

El dataset original contiene valores nulos y columnas poco √∫tiles.  
Acciones realizadas:  

1. **Eliminar columnas con demasiados valores nulos** ‚Üí se elimin√≥ `Cabin`.  
2. **Eliminar filas con valores nulos en `Embarked`** ‚Üí eran pocas, as√≠ que no afectan mucho.  
3. **Imputar valores faltantes en `Age`** ‚Üí se reemplazaron con la mediana, ya que es m√°s robusta frente a valores extremos que la media.  

Tras la limpieza:  
- La variable `Age` ya no contiene nulos.  
- La variable `Embarked` solo tiene valores v√°lidos.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lectura de archivo
df = pd.read_csv("PROYECTOS/tested.csv")

# Mostr√°r las primeras filas para visualizar datos
df.head()

#Estructura del data set
df.info()

#Estadistica num√©rica
df.describe()

#Revisi√≥n de valores faltantes
df.isnull().sum()

#Distribuci√≥n de variables categ√≥ricas
df['Sex'].value_counts()
df['Pclass'].value_counts()

#Estilo visual para gr√°ficos
sns.set(style="whitegrid")

# Gr√°fico de barras por clases (Pclass)
sns.countplot(data=df, x='Pclass', palette='viridis')
plt.title('Distribuci√≥n de Pasajeros por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad de pasajeros')
plt.show()


# Gr√°fico de barras por genero (Sex)
sns.countplot(x='Sex', data=df , palette='pastel')
plt.title('Distribuci√≥n de pasajeros por sexo')
plt.xlabel("sexo")
plt.ylabel("N√∫mero de pasajeros")
plt.show()

# Gr√°fico de barras de supervivencia (Survived)
sns.countplot(x='Survived', data=df , palette='Set2')
plt.title('Distribuci√≥n de pasajeros por supervivencia')
plt.xlabel("0 = No sobrevivi√≥, 1 = Sobrevivi√≥")
plt.ylabel("N√∫mero de pasajeros")
plt.show()

# Gr√°fico de barras Supervivencia por sexo
sns.countplot(x='Survived', hue='Sex', data=df , palette='Set2')
plt.title('Distribuci√≥n de pasajeros por supervivencia por sexo')
plt.xlabel("Sexo")
plt.xlabel("N√∫mero de pasajeros")
plt.ylabel("Sobrevivi√≥")
plt.show()

# Gr√°fico de barras Supervivencia por clase
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set3')
plt.title('Supervivencia por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad de pasajeros')
plt.legend(title='Sobrevivi√≥')
plt.show()

#Histograma: Distribuci√≥n de edades
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True, color='skyblue')
plt.title("Distribuci√≥n de Edades")
plt.xlabel("Edad")
plt.ylabel("Distribuci√≥n de Edades de los Pasajeros del Titanic")
plt.show()

#Boxplot: Edad por clase
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Pclass', y='Age', palette='Accent')
plt.title("Distribuci√≥n por Clase")
plt.xlabel("Clase")
plt.ylabel("Edad")
plt.show()

# Copia del df para limpieza
df_clean = df.copy()

# Revisar valores nulos
print(df_clean.isnull().sum())

# Porcentaje de valores nulos por columna
porcentaje_nulos = df_clean.isnull().mean() * 100
print("Porcentaje de valores nulos por columna:\n", porcentaje_nulos)

# Eliminar columnas con demasiados valores nulos
df_clean.drop(columns=["Cabin"], inplace=True)

# Eliminar filas con valores nulos en 'Embarked'
df_clean.dropna(subset=["Embarked"], inplace=True)

# Imputar edad faltante con la mediana
edad_mediana = df_clean['Age'].median()
df_clean['Age'].fillna(edad_mediana, inplace=True)

# Verificar valores nulos nuevamente
print("Valores nulos tras la limpieza:\n", df_clean.isnull().sum())

# Confirmar el tipo de datos
print("\nTipos de datos despu√©s de limpieza:")
print(df_clean.dtypes)


```
